```python
import pandas as pd
```

```python
df=pd.read_csv('EPL_Set.csv')
df.head()
```

```python
df.describe()
```

```python
df = df.dropna()
```

```python
#to predict FTR using HTR HTAG HTHG FTHG HTAG
y=df.FTR
df_features = ['FTHG','FTAG','HTAG','HTHG']
X=df[df_features]
X.head()
```

```python
X.dropna()
```

```python
from sklearn.tree import DecisionTreeRegressor
df_model=DecisionTreeRegressor(random_state=1)
```

```python
df['FTR']=df['FTR'].replace(to_replace = 'H',value = 1.0)
df['FTR']=df['FTR'].replace(to_replace = 'D',value = 0.0)
df['FTR']=df['FTR'].replace('A',2.0)
df['HTR']=df['HTR'].replace(to_replace = 'H',value = 1.0)
df['HTR']=df['HTR'].replace(to_replace = 'D',value = 0.0)
df['HTR']=df['HTR'].replace('A',2.0)
df
```

```python
df_model.fit(X,y)
```

```python
print("Making predictions for the following:")
print(X.head())
print("The predictions are")
print(df_model.predict(X.head()))
```
